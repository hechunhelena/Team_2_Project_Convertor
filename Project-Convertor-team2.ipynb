## Project- Convertor

#### Describle of project: Create a program that can be used to convert temperature, length, weight, pressure.

```python
# Define a function to convert Fahrenheit to Celsius
def fahrenheit_to_celsius(fahrenheit):
    return (fahrenheit - 32) * 5/9

# Define a function to convert Celsius to Fahrenheit
def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32

# Define a function to convert miles to kilometers
def miles_to_km(miles):
    return miles * 1.60934

# Define a function to convert kilometers to miles
def km_to_miles(km):
    return km / 1.60934

# Define a function to convert pounds to kilograms
def pound_to_kilograms(pound):
    return pound * 0.453592

# Define a function to convert kilograms to pounds
def kg_to_pound(kg):
    return kg / 0.453592

# Display the conversion menu
while True:
    print("\nChoose a conversion option:")
    print("1. Fahrenheit to Celsius")
    print("2. Celsius to Fahrenheit")
    print("3. Miles to Kilometers")
    print("4. Kilometers to Miles")
    print("5. Pounds to Kilograms")
    print("6. Kilograms to Pounds")
    print("7. Exit")
    
    # Get user's choice
    choice = input("Enter your choice (1/2/3/4/5/6/7): ")

    # Perform the selected conversion or exit
    if choice == '1':
        fahrenheit = float(input("Enter temperature in Fahrenheit: "))
        celsius = fahrenheit_to_celsius(fahrenheit)
        print(f"{fahrenheit} Fahrenheit is equal to {celsius} Celsius.")
    elif choice == '2':
        celsius = float(input("Enter temperature in Celsius: "))
        fahrenheit = celsius_to_fahrenheit(celsius)
        print(f"{celsius} Celsius is equal to {fahrenheit} Fahrenheit.")
    elif choice == '3':
        miles = float(input("Enter length in miles: "))
        km = miles_to_km(miles)
        print(f"{miles} miles is equal to {km} kilometers.")
    elif choice == '4':
        km = float(input("Enter length in kilometers: "))
        miles = km_to_miles(km)
        print(f"{km} kilometers is equal to {miles} miles.")
    elif choice == '5':
        pound = float(input("Enter weight in pounds: "))
        kg = pound_to_kilograms(pound)
        print(f"{pound} pounds is equal to {kg} kilograms.")
    elif choice == '6':
        kg = float(input("Enter weight in kilograms: "))
        pound = kg_to_pound(kg)
        print(f"{kg} kilograms is equal to {pound} pounds.")
    elif choice == '7':
        print("Goodbye!")
        break
    else:
        print("Invalid choice. Please enter a valid option (1/2/3/4/5/6/7).")
```
